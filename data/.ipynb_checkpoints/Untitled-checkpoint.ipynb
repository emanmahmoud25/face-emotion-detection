```python
import zipfile
import os

zip_path = "face-expression-recognition-dataset.zip"
extract_to = "data"

# create output folder if it doesn't exist
os.makedirs(extract_to, exist_ok=True)

# extract zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("‚úÖ Unzipped successfully to:", extract_to)
```

```python
import os

base_dir = "images"  # ÿ∫Ÿäÿ±Ÿä ÿ≠ÿ≥ÿ® ŸÖŸÉÿßŸÜ ÿßŸÑÿµŸàÿ± ÿßŸÑŸÅÿπŸÑŸä

folders = ['train', 'validation']

for folder in folders:
    path = os.path.join(base_dir, folder)
    if not os.path.exists(path):
        print(f"‚ö†Ô∏è Folder not found: {path}")
        continue

    print(f"\nüìÅ Folder: {folder}")
    
    for category in os.listdir(path):
        category_path = os.path.join(path, category)
        if os.path.isdir(category_path):
            count = len([
                f for f in os.listdir(category_path)
                if os.path.isfile(os.path.join(category_path, f))
            ])
            print(f"  - {category}: {count} images")
```
